## Visualization

### 1. Price Variation Across Locations (Downtown vs Suburban)

```python
top_localities = df.groupby('Locality')['Price'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(12, 6))
ax = sns.barplot(x=top_localities.values, y=top_localities.index, palette="viridis")
plt.title("Average Price in Top 10 Localities", fontsize=16)
plt.xlabel("Average Price (in USD)", fontsize=12)
plt.ylabel("Localities", fontsize=12)

# Add labels to each bar
for index, value in enumerate(top_localities.values):
    plt.text(value, index, f"${value:,.0f}", va='center', fontsize=10)

plt.tight_layout()
plt.show() 
```

### 2. Effect of Size on Price (Properties above 3000 Sq.Ft.)

```python

large_properties = df[df['Area (Sq.Ft.)'] > 3000]
plt.figure(figsize=(10, 6))
sns.scatterplot(data=large_properties, x='Area (Sq.Ft.)', y='Price', hue='City', palette='tab10', alpha=0.7)
plt.title("Effect of Size on Price (Properties > 3000 Sq.Ft.)", fontsize=16)
plt.xlabel("Area (Sq.Ft.)", fontsize=12)
plt.ylabel("Price (in USD)", fontsize=12)
plt.legend(title="City", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

### 3. Correlation Between Features

```python
# Analyze how numerical features such as Area (Sq.Ft.), Price, and Number of Rooms are related.
correlation_data = df[['Price', 'Area (Sq.Ft.)', 'Number of Rooms', 'Number of Bathroom', 'Price/Sq.Ft.']].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_data, annot=True, cmap="YlGnBu", fmt=".2f")
plt.title("Correlation Between Features", fontsize=16)
plt.tight_layout()
plt.show() 
```

### 4. City-Wise Availability of Properties

```python
# City-wise number of listings
city_listings = df['City'].value_counts().head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x=city_listings.values, y=city_listings.index, palette="Blues")
plt.title("Top 10 Cities by Number of Property Listings", fontsize=16)
# Add labels to each bar
for index, value in enumerate(city_listings.values):
    plt.text(value, index, f"{value:,.0f}", va='center', fontsize=10)
    
plt.xlabel("Number of Listings", fontsize=12)
plt.ylabel("Cities", fontsize=12)
plt.tight_layout()
plt.show() 
```

### 5. Price per Square Foot Analysis

```python
# Top 10 cities by average price per square foot
price_per_sqft = df.groupby('City')['Price/Sq.Ft.'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=price_per_sqft.values, y=price_per_sqft.index, palette="magma")
plt.title("Top 10 Cities by Price per Square Foot", fontsize=16)

# Add labels to each bar
for index, value in enumerate(price_per_sqft.values):
    plt.text(value, index, f"{value:,.0f}", va='center', fontsize=10)
    
plt.xlabel("Average Price/Sq.Ft. (in USD)", fontsize=12)
plt.ylabel("Cities", fontsize=12)
plt.tight_layout()
plt.show() 
```

### 6. Impact of Furnishing Status on Prices

```python
# Average price by furnishing status
furnishing_prices = df.groupby('Furnishing Status')['Price'].mean().sort_values()

plt.figure(figsize=(8, 6))
sns.barplot(x=furnishing_prices.values, y=furnishing_prices.index, palette="Set2")
# Add labels to each bar
for index, value in enumerate(furnishing_prices.values):
    plt.text(value, index, f"${value:,.0f}", va='center', fontsize=10)
    
plt.title("Impact of Furnishing Status on Price", fontsize=16)
plt.xlabel("Average Price (in USD)", fontsize=12)
plt.ylabel("Furnishing Status", fontsize=12)
plt.tight_layout()
plt.show() 
```
