```python
import pandas as pd

# Assuming data_cleaned is your DataFrame
data_cleaned['info'] = data_cleaned['info'].str.replace('Property Information\n', '', regex=False)

# Define the column names based on the known labels
columns = [
    "City", "Locality", "Listing Type", "Property Type", "Area", "Price", "Possession Status",
    "Possession Date", "Furnishing Status", "Age of Property (Years)", "Number of Rooms",
    "Number of Bathroom", "Covered Parking", "Open/Uncovered Parking", "Balcony",
    "Power Back-up", "Facing", "Flooring"
]

# Initialize an empty DataFrame to store the results
split_data = pd.DataFrame(columns=columns)

# Iterate over each row in the DataFrame
for idx, row in data_cleaned.iterrows():
    # Check if 'info' is a string to handle missing or NaN values
    if isinstance(row['info'], str):
        # Split the row values by '\n' to get key-value pairs
        property_info = row['info'].split('\n')
        
        # Use a for loop to assign the values to the corresponding columns
        data_dict = {}
        for i in range(0, len(property_info), 2):  # Iterate over the key-value pairs
            key = property_info[i].strip()
            if i + 1 < len(property_info):  # Check if there is a corresponding value
                value = property_info[i + 1].strip()
                if key in columns:
                    data_dict[key] = value
        
        # Append the dictionary to the DataFrame
        split_data = split_data.append(data_dict, ignore_index=True)
    else:
        # If 'info' is NaN, append an empty row
        split_data = split_data.append({}, ignore_index=True)

# Display the cleaned data
print(split_data)

# # Save the cleaned data to a new CSV file if needed
# output_path = '/mnt/data/cleaned_property_data.csv'
# split_data.to_csv(output_path, index=False)
# print(f"Cleaned data saved to {output_path}")
```

```python
# Step 1: Clean 'Price' column by converting to numerical format
def convert_price(price_str):
    if isinstance(price_str, str):  # Check if price_str is a string
        # Convert "L" (lakhs) and "Cr." (crores) to numerical value in rupees
        if 'L' in price_str:
            return float(price_str.replace("L", "").strip()) * 1e5  # 1 Lakh = 100,000
        elif 'Cr' in price_str:
            return float(price_str.replace("Cr.", "").strip()) * 1e7  # 1 Crore = 10,000,000
    return None  # Return None if price_str is not a string

# Apply the conversion function to the 'Price' column
data_cleaned['Price'] = data_cleaned['Price'].apply(convert_price)

# Display the cleaned data to verify the conversion
print(data_cleaned[['Price']].head())
```

```python
# If the amenities column is confirmed as 'aminities', proceed with the following steps:
if 'aminities' in data_cleaned.columns:
    # Step 2: Ensure the 'aminities' column has strings and fill missing values
    data_cleaned['aminities'] = data_cleaned['aminities'].astype(str).fillna('')

    # Step 3: Extract unique amenities by splitting and normalizing the text
    unique_amenities = set()
    for amenities in data_cleaned['aminities']:
        for amenity in amenities.split(','):
            unique_amenities.add(amenity.strip())  # Strip extra whitespace

    # Step 4: Create binary columns for each unique amenity
    for amenity in unique_amenities:
        data_cleaned[amenity] = data_cleaned['aminities'].apply(lambda x: 1 if amenity in x else 0)

    # Step 5: Drop the original 'aminities' column if no longer needed
    df_cleaned = data_cleaned.drop(columns=['aminities'])

    # Display the first few rows of the transformed dataset
    print(df_cleaned.head())
else:
    print("The 'aminities' column was not found in the dataset. Please check the column names.")
```

```python
import re

# Define a function to clean column names
def clean_column_names(columns):
    cleaned_columns = []
    for col in columns:
        # Remove numbers and special characters, replace with space
        clean_name = re.sub(r'[^a-zA-Z\s]', '', col)
        # Replace multiple spaces with a single space, then strip leading/trailing spaces
        clean_name = re.sub(r'\s+', ' ', clean_name).strip()
        # Convert to title case to ensure consistent formatting
        cleaned_columns.append(clean_name.title())
    return cleaned_columns

# Apply cleaning function to column names
data_cleaned.columns = clean_column_names(data_cleaned.columns)

# Display cleaned column names
print("Cleaned Column Names:", data_cleaned.columns)
```

```python
def check_and_drop_duplicates(df_filtered):
    # Check for duplicate rows
    num_duplicates = df_filtered.duplicated().sum()
    
    if num_duplicates > 0:
        print(f"Found {num_duplicates} duplicate rows. Dropping them now.")
        # Drop duplicate rows
        df_filtered = df_filtered.drop_duplicates()
        print("Duplicates removed successfully.")
    else:
        print("No duplicate rows found.")
    
    return df_filtered

# Assuming df_filtered is your DataFrame, call the function
df_filtered = check_and_drop_duplicates(df_filtered)

# Verify the result
print(df_filtered.head())
```

```python
import pandas as pd

# Function to convert area to Sq.Ft.
def convert_to_sqft(area_str):
    # Extract number and unit
    number = float(area_str.split()[0])
    unit = area_str.split()[1]
    
    if unit == 'Sq.Ft.':
        return number
    elif unit == 'Sq.Yd.':
        return number * 9  # 1 Sq.Yd. = 9 Sq.Ft.
    elif unit == 'Acres':
        return number * 43560  # 1 Acre = 43560 Sq.Ft.
    elif unit == 'Sq.Mt.':
        return number * 10.7639  # 1 Sq.Mt. = 10.7639 Sq.Ft.
    # Add more units as needed
    return None

# Apply the conversion function to the Area column
df_filtered['Area (Sq.Ft.)'] = df_filtered['Area'].apply(convert_to_sqft)

# Display the DataFrame with the converted areas
print(df_filtered)
```
